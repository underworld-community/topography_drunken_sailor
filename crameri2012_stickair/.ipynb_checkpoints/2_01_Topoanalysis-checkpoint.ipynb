```python
import h5py   
import matplotlib.pyplot as plt
```

```python
def load_h5_tracer(fdir,tracer_name,number):
    """
    load the data of the passive tracer from h5 file
    -----------
    Input:
    fdir   : *Str*, input file dir 
    tracer_name : *Str*, name of tracer
    number : *Int*, the number of file
    -----------
    Output:
    coord : *Np.array*
    data  : *Np.array*  
    label : *str*
    """
    fname = fdir+tracer_name +"-"+str(number)+".h5"
    fh5   = h5py.File(fname ,'r')  
    fdata = fh5["data"][()]
    coord = fdata[:,0]
    data  = fdata[:,1]
    return coord,data #label
```

```python
### Different boundary conditions at bottom
tracer_name = "Surface"
number = 4
dT = 1

fig, ax = plt.subplots(1, 1, figsize=(12,3))
ax.set(xlabel='Distance [km]', ylabel='Topography [m]') #,xticks=xticks_fig,yticks=yticks_fig,title=fname_fig)

fdir = "OPC2012_uwg_airdensity_5k/"
label = "Air density"
coord,elev = load_h5_tracer(fdir,tracer_name,number)
ax.plot(coord,elev*1000,label=label)

fdir = "OPC2012_uwg_waterdensity_5k/"
label = "Water density"
coord,elev = load_h5_tracer(fdir,tracer_name,number)
ax.plot(coord,elev*1000,label=label)


fdir ="OPC2012_uwg_waterdensity_rm_5k/"
label = "Water density rm_ref"
coord,elev = load_h5_tracer(fdir,tracer_name,number)
ax.plot(coord,elev*1000,label=label)

fdir ="OPC2012_uwg_airdensity_rm_5k/"
label = "Air density rm_ref"
coord,elev = load_h5_tracer(fdir,tracer_name,number)
ax.plot(coord,elev*1000,label=label)

ax.set_xlim([0,2800])
#ax.set_ylim([-200,1200])
ax.legend(loc = 'upper right',prop = {'size':8}) 
ax.grid()

fname = "Topography at "+ str(int(number*dT)) + " Ma"
plt.savefig(fname)
```

```python
tracer_name = "Surface"
number = 4
dT = 1

fig, (ax1,ax2) = plt.subplots(2,1, figsize=(6,6))
ax.set(xlabel='Distance [km]', ylabel='Topography [m]') #,xticks=xticks_fig,yticks=yticks_fig,title=fname_fig)

fdir = "OPC2012_uwg_airdensity_5k/"
title = "Air density"
label1 = "0-1400 km"
label2 ="1400-2800 km"

coord,elev = load_h5_tracer(fdir,tracer_name,number)
ax1.plot(coord[:2500],elev[:2500]*1000,label=label1)
ax1.plot(coord[:2500],elev[2500:][::-1]*1000,label=label2)
ax1.set_xlim([0,1400])
ax1.set_ylim([-200,600])
ax1.legend(loc = 'upper left',prop = {'size':8}) 
ax1.grid()
ax1.set_aspect(1/2)
ax1.set_title(title)

fdir = "OPC2012_uwg_waterdensity_5k/"
title = "Water density"
coord,elev = load_h5_tracer(fdir,tracer_name,number)
ax2.plot(coord[:2500],elev[:2500]*1000,label=label1)
ax2.plot(coord[:2500],elev[2500:][::-1]*1000,label=label2)
ax2.set_xlim([0,1400])
ax2.set_ylim([-200,600])
ax2.legend(loc = 'upper left',prop = {'size':8}) 
ax2.grid()
ax2.set_aspect(1/2)
ax2.set_title(title)

fname = "Symmetry of the topography at "+ str(int(number*dT)) + " Ma"
plt.savefig(fname)
```
